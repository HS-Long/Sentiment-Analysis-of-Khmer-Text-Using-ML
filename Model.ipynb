## Step 3: Khmer-Specific Text Preprocessing

### Key Task: Apply Khmer-specific preprocessing (Unicode normalization, slang handling)

Khmer text presents unique challenges:
1. **Unicode Normalization (NFC)**: Khmer characters can have multiple Unicode representations
2. **Informal Writing & Slang**: Social media contains non-standard Khmer expressions
3. **Mixed Scripts**: Khmer text often includes English words, emojis, and special characters
4. **Special Markers**: URLs, symbols, and formatting characters need to be removed

Our preprocessing pipeline includes:
- Unicode normalization to standardize Khmer characters
- Khmer slang mapping to formal language
- Removal of URLs, emojis, and special symbols
- Preserving Khmer Unicode range (U+1780 to U+17FF)

````python
import re
import unicodedata

# Common Khmer slang/informal mapping (expand this with more slangs)
KHMER_SLANG = {
    "មិនចេះ": "មិនដឹង",
    "ចេះតែ": "តែងតែ",
    "អត់": "មិន",
    "ហ៊ាន": "ហ៊ាន",
    # Add more Khmer slang mappings as you discover them
}

def preprocess_khmer(text):
    # Unicode normalization (NFC for Khmer)
    text = unicodedata.normalize("NFC", text)
    
    # Lowercase (safe for Khmer + English mix)
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)
    
    # Remove special markers like :://, //, etc.
    text = re.sub(r"[:]+\/\/|\/\/", " ", text)
    
    # Handle Khmer slang (replace with standard form)
    for slang, standard in KHMER_SLANG.items():
        text = text.replace(slang, standard)
    
    # Remove emojis & symbols (keep Khmer + numbers + basic punctuation)
    text = re.sub(r"[^\u1780-\u17FF0-9\s\?\!\.]", " ", text)
    
    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    
    return text
````

## Step 4: Feature Extraction

### TF-IDF Vectorization

Convert cleaned Khmer text into numerical features using TF-IDF (Term Frequency-Inverse Document Frequency):
- **N-grams**: Using unigrams and bigrams (1,2) to capture word context
- **max_df=0.9**: Ignore terms that appear in more than 90% of documents
- **min_df=5**: Ignore terms that appear in fewer than 5 documents
- **Analyzer='word'**: Word-level tokenization

This vectorization approach helps capture semantic meaning in Khmer text despite the lack of Khmer-specific NLP tools.

````python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    analyzer="word",
    ngram_range=(1, 2),     # unigram + bigram
    max_df=0.9,
    min_df=5
)
````

### Model 1: Logistic Regression

Train Logistic Regression with hyperparameter tuning:
- **C**: Regularization strength (0.01 to 10)
- **Solver**: Optimization algorithm (lbfgs, saga)
- **max_iter=2000**: Sufficient iterations for convergence

````python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV

pipe_lr = Pipeline([
    ("tfidf", tfidf),
    ("clf", LogisticRegression(
        class_weight=class_weight,
        max_iter=2000
    ))
])

param_lr = {
    "clf__C": [0.01, 0.1, 1, 5, 10],
    "clf__solver": ["lbfgs", "saga"]
}

rs_lr = RandomizedSearchCV(
    pipe_lr,
    param_lr,
    n_iter=10,
    scoring="f1_macro",
    cv=3,
    n_jobs=-1,
    random_state=42
)

rs_lr.fit(X_train, y_train)
````

### Model 2: Support Vector Machine (SVM)

Train LinearSVC with hyperparameter tuning:
- **C**: Regularization parameter (0.01 to 10)
- **class_weight='balanced'**: Handle class imbalance
- LinearSVC is faster than kernel SVM for high-dimensional text data

````python
from sklearn.svm import LinearSVC

pipe_svm = Pipeline([
    ("tfidf", tfidf),
    ("clf", LinearSVC(
        class_weight=class_weight
    ))
])

param_svm = {
    "clf__C": [0.01, 0.1, 1, 5, 10]
}

rs_svm = RandomizedSearchCV(
    pipe_svm,
    param_svm,
    n_iter=10,
    scoring="f1_macro",
    cv=3,
    n_jobs=-1,
    random_state=42
)

rs_svm.fit(X_train, y_train)
````

### Model 3: Multinomial Naive Bayes

Train Naive Bayes with hyperparameter tuning:
- **alpha**: Smoothing parameter (0.01 to 1.0)
- Fast training and works well with TF-IDF features
- Assumes feature independence (words in text)

````python
from sklearn.naive_bayes import MultinomialNB

pipe_nb = Pipeline([
    ("tfidf", tfidf),
    ("clf", MultinomialNB())
])

param_nb = {
    "clf__alpha": [0.01, 0.1, 0.5, 1.0]
}

rs_nb = RandomizedSearchCV(
    pipe_nb,
    param_nb,
    n_iter=5,
    scoring="f1_macro",
    cv=3,
    n_jobs=-1,
    random_state=42
)

rs_nb.fit(X_train, y_train)
````

````python
try:
    import tensorflow as tf
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Embedding, LSTM, Bidirectional, Dense, Dropout
    from tensorflow.keras.preprocessing.text import Tokenizer
    from tensorflow.keras.preprocessing.sequence import pad_sequences
    from tensorflow.keras.callbacks import EarlyStopping
    
    # Tokenize text for LSTM
    max_words = 5000
    max_len = 100
    
    tokenizer = Tokenizer(num_words=max_words, oov_token='<OOV>')
    tokenizer.fit_on_texts(X_train)
    
    X_train_seq = tokenizer.texts_to_sequences(X_train)
    X_test_seq = tokenizer.texts_to_sequences(X_test)
    
    X_train_pad = pad_sequences(X_train_seq, maxlen=max_len, padding='post')
    X_test_pad = pad_sequences(X_test_seq, maxlen=max_len, padding='post')
    
    # Encode labels
    le_lstm = LabelEncoder()
    y_train_lstm = le_lstm.fit_transform(y_train)
    y_test_lstm = le_lstm.transform(y_test)
    
    # Build BiLSTM model
    model_lstm = Sequential([
        Embedding(input_dim=max_words, output_dim=128, input_length=max_len),
        Bidirectional(LSTM(64, return_sequences=True)),
        Dropout(0.5),
        Bidirectional(LSTM(32)),
        Dropout(0.5),
        Dense(32, activation='relu'),
        Dropout(0.3),
        Dense(3, activation='softmax')
    ])
    
    model_lstm.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    
    # Train with early stopping
    early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
    
    print("Training Bidirectional LSTM...")
    history = model_lstm.fit(
        X_train_pad, y_train_lstm,
        epochs=20,
        batch_size=32,
        validation_split=0.2,
        callbacks=[early_stop],
        verbose=1
    )
    
    print("✓ BiLSTM trained successfully")
    
except ImportError as e:
    print(f"⚠ TensorFlow not installed. Run: pip install tensorflow")
    print(f"Error: {e}")
    model_lstm = None
    tokenizer = None
````

## Step 7: Challenges and Observations in Khmer Sentiment Classification

### Challenges Specific to Khmer Text:

1. **Limited NLP Resources**: 
   - Unlike English, Khmer lacks extensive pre-trained models and linguistic tools
   - No standard tokenizers or POS taggers for Khmer
   - Solution: Unicode-aware character-level and word-level features with TF-IDF

2. **Unicode Complexity**:
   - Khmer has complex Unicode structure with combining characters
   - Multiple representations of the same character
   - Solution: NFC normalization to standardize representation

3. **Informal Writing Styles**:
   - Social media uses heavy slang, abbreviations, and mixed scripts
   - Non-standard spelling variations
   - Solution: Slang mapping dictionary (needs continuous expansion)

4. **Context-Dependent Sentiment**:
   - Khmer sentiment often relies on cultural context and implied meaning
   - Sarcasm and irony are common in Khmer social media
   - Solution: N-gram features to capture some contextual information

5. **Class Imbalance**:
   - Real-world data often has unbalanced sentiment distribution
   - Solution: Class weights and stratified sampling

6. **Mixed Language Content**:
   - Khmer posts often include English words and emojis
   - Solution: Preserve both Khmer and numbers in preprocessing

## 📊 ROC Curve Analysis and Optimal Threshold Finding

This section analyzes ROC curves for each class and finds optimal classification thresholds using multiple methods:
- **Youden's J Statistic**: Maximizes (TPR - FPR)
- **F1 Score Optimization**: Maximizes F1 score
- **Cost-Sensitive Optimization**: Minimizes weighted misclassification cost

````python
# Function to compute ROC curves for multi-class classification
def compute_roc_curves_multiclass(y_true, y_proba, classes):
    """
    Compute ROC curves for multi-class classification using One-vs-Rest approach.
    
    Parameters:
    - y_true: True labels
    - y_proba: Predicted probabilities for each class
    - classes: List of class labels
    
    Returns:
    - Dictionary containing FPR, TPR, thresholds, and AUC for each class
    """
    # Binarize the labels
    y_true_bin = label_binarize(y_true, classes=classes)
    
    roc_data = {}
    for i, class_name in enumerate(classes):
        fpr, tpr, thresholds = roc_curve(y_true_bin[:, i], y_proba[:, i])
        roc_auc = auc(fpr, tpr)
        
        roc_data[class_name] = {
            'fpr': fpr,
            'tpr': tpr,
            'thresholds': thresholds,
            'auc': roc_auc
        }
    
    return roc_data

# Compute ROC curves for the best model
print("Computing ROC curves for the best model...")
y_proba = best_model_obj.predict_proba(X_test)
roc_data = compute_roc_curves_multiclass(y_test, y_proba, classes)

print("\nROC AUC Scores by Class:")
for class_name, data in roc_data.items():
    print(f"  {class_name}: {data['auc']:.4f}")
````

````python
# Function to find optimal threshold using Youden's J statistic
def find_optimal_threshold_youden(fpr, tpr, thresholds):
    """
    Find optimal threshold using Youden's J statistic (maximizes TPR - FPR).
    
    Parameters:
    - fpr: False Positive Rate array
    - tpr: True Positive Rate array
    - thresholds: Threshold array
    
    Returns:
    - Optimal threshold value
    """
    j_scores = tpr - fpr
    optimal_idx = np.argmax(j_scores)
    optimal_threshold = thresholds[optimal_idx]
    return optimal_threshold, j_scores[optimal_idx]

# Function to find optimal threshold using F1 score
def find_optimal_threshold_f1(y_true, y_proba, class_idx):
    """
    Find optimal threshold by maximizing F1 score.
    
    Parameters:
    - y_true: True labels (binarized for the class)
    - y_proba: Predicted probabilities for the class
    - class_idx: Index of the class
    
    Returns:
    - Optimal threshold value and max F1 score
    """
    from sklearn.metrics import f1_score
    
    # Test different thresholds
    thresholds = np.linspace(0, 1, 100)
    f1_scores = []
    
    for threshold in thresholds:
        y_pred = (y_proba >= threshold).astype(int)
        f1 = f1_score(y_true, y_pred, zero_division=0)
        f1_scores.append(f1)
    
    optimal_idx = np.argmax(f1_scores)
    optimal_threshold = thresholds[optimal_idx]
    max_f1 = f1_scores[optimal_idx]
    
    return optimal_threshold, max_f1, f1_scores

# Find optimal thresholds for all classes
print("\nFinding Optimal Thresholds...\n")
optimal_thresholds = {}

for class_name in classes:
    print(f"Class: {class_name}")
    
    # Youden's J method
    fpr = roc_data[class_name]['fpr']
    tpr = roc_data[class_name]['tpr']
    thresholds = roc_data[class_name]['thresholds']
    
    optimal_thresh_youden, j_score = find_optimal_threshold_youden(fpr, tpr, thresholds)
    print(f"  Youden's J Statistic: Threshold = {optimal_thresh_youden:.4f}, J = {j_score:.4f}")
    
    # F1 score method
    class_idx = list(classes).index(class_name)
    y_true_bin = label_binarize(y_test, classes=classes)[:, class_idx]
    y_proba_class = y_proba[:, class_idx]
    
    optimal_thresh_f1, max_f1, f1_scores = find_optimal_threshold_f1(y_true_bin, y_proba_class, class_idx)
    print(f"  F1 Score Method: Threshold = {optimal_thresh_f1:.4f}, F1 = {max_f1:.4f}")
    
    optimal_thresholds[class_name] = {
        'youden': optimal_thresh_youden,
        'f1': optimal_thresh_f1,
        'default': 0.5,
        'auc': roc_data[class_name]['auc']
    }
    print()
````

````python
# Compare performance with different thresholds
from sklearn.metrics import classification_report

def predict_with_custom_thresholds(model, X, threshold_dict, classes, method='f1'):
    """
    Make predictions using custom thresholds for each class.
    
    Parameters:
    - model: Trained model
    - X: Features
    - threshold_dict: Dictionary of optimal thresholds
    - classes: List of class labels
    - method: 'f1', 'youden', or 'default'
    
    Returns:
    - Predicted labels
    """
    y_proba = model.predict_proba(X)
    predictions = []
    
    for proba_row in y_proba:
        scores = []
        for i, class_name in enumerate(classes):
            threshold = threshold_dict[class_name][method]
            # Adjust probability based on threshold
            adjusted_score = proba_row[i] / threshold if threshold > 0 else 0
            scores.append(adjusted_score)
        
        # Predict the class with highest adjusted score
        pred_idx = np.argmax(scores)
        predictions.append(classes[pred_idx])
    
    return np.array(predictions)

# Test with default threshold (0.5)
print("\n" + "="*80)
print("🔵 PERFORMANCE WITH DEFAULT THRESHOLD (0.5)")
print("="*80)
y_pred_default = best_model_obj.predict(X_test)
print(classification_report(y_test, y_pred_default, target_names=classes))

# Test with Youden's J threshold
print("\n" + "="*80)
print("🟠 PERFORMANCE WITH YOUDEN'S J OPTIMAL THRESHOLDS")
print("="*80)
y_pred_youden = predict_with_custom_thresholds(best_model_obj, X_test, optimal_thresholds, classes, method='youden')
print(classification_report(y_test, y_pred_youden, target_names=classes))

# Test with F1 optimal threshold
print("\n" + "="*80)
print("🟢 PERFORMANCE WITH F1 OPTIMAL THRESHOLDS")
print("="*80)
y_pred_f1 = predict_with_custom_thresholds(best_model_obj, X_test, optimal_thresholds, classes, method='f1')
print(classification_report(y_test, y_pred_f1, target_names=classes))
````

### 🎯 Key Findings and Recommendations

**Understanding Optimal Thresholds:**

1. **Default Threshold (0.5)**: Standard approach - predicts the class with highest probability above 0.5
2. **Youden's J Statistic**: Maximizes the distance from the ROC curve to the diagonal (TPR - FPR)
   - Best for balanced sensitivity and specificity
3. **F1 Optimal**: Maximizes the F1 score (harmonic mean of precision and recall)
   - Best for balanced precision and recall
   
**How to Use Optimal Thresholds in Production:**

```python
# Example: Making predictions with optimal F1 thresholds
y_proba = model.predict_proba(X_new)
predictions = predict_with_custom_thresholds(model, X_new, optimal_thresholds, classes, method='f1')
```

**When to Use Each Method:**
- **F1 Method**: When you want balanced precision and recall (recommended for most cases)
- **Youden Method**: When false positives and false negatives have equal cost
- **Default (0.5)**: When computational simplicity is prioritized over optimal performance
